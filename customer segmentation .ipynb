# K-mean Algorithrm


use elbow method to find optimal clusters

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    # Assuming you have data to fit, e.g., km.fit(data)
    # Replace 'data' with your actual data
    km.fit(x)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method', fontsize=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Sample data (replace this with your actual data)
x = np.random.rand(200, 2)

plt.style.use('fivethirtyeight')

km = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s=100, c='Pink', label='miser')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s=100, c='Yellow', label='general')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s=100, c='Cyan', label='target')
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s=100, c='Magenta', label='spendthrift')
plt.scatter(x[y_means == 4, 0], x[y_means == 4, 1], s=100, c='Orange', label='careful')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=50, c='Blue', label='centroid')

plt.title('KMeans Clustering between Annual Income and Spending Score', fontsize=20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    # Assuming you have data to fit, e.g., km.fit(data)
    # Replace 'data' with your actual data
    km.fit(y)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method', fontsize=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Sample data (replace this with your actual data)
x = np.random.rand(200, 2)

plt.style.use('fivethirtyeight')

km = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
z_means = km.fit_predict(y)

plt.scatter(y[z_means == 0, 0], y[z_means == 0, 1], s=100, c='Pink', label='Unusual customers')
plt.scatter(y[z_means == 1, 0], y[z_means == 1, 1], s=100, c='Orange', label='priority customers')
plt.scatter(y[z_means == 2, 0], y[z_means == 2, 1], s=100, c='Lightgreen', label='target customers(young)')
plt.scatter(y[z_means == 3, 0], y[z_means == 3, 1], s=100, c='Red', label='target customers(young)')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=50, c='Black', label='centroid')

plt.title('KMeans Clustering between Age and Spending Score', fontsize=20)
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()

plt.show()
```
